```python
# displaying a single grid
def DisplayGrid(grid):
    grid = np.asarray(grid)
    nrows = len(grid[:,0])
    ncols = len(grid[0,:])
    height, width = nrows*50, ncols*50
    image = Image.new(size=(width,height),mode='RGB',color=(255,255,255))
    draw = ImageDraw.Draw(image)
    r = 0 
    for row in grid:
        c = 0
        for col in row:
            draw.rectangle(xy=[c*50,r*50,(c+1)*50,(r+1)*50], fill=colorMap[np.abs(grid[r][c])])
            c += 1
        r += 1
    for i in range(ncols):
        draw.line([(i+1)*50,0,(i+1)*50,height],fill="grey")
    for i in range(nrows):
        draw.line([0,(i+1)*50,width,(i+1)*50],fill="grey")
    display(image)
```

```python
# displaying two grids side by side
def DisplayGrids(grid1,grid2):
    # Grid 1
    grid1 = np.asarray(grid1)
    nrows = len(grid1[:,0])
    ncols = len(grid1[0,:])
    height, width = nrows*50, ncols*50
    image1 = Image.new(size=(width,height),mode='RGB',color=(255,255,255))
    draw = ImageDraw.Draw(image1)
    r = 0 
    for row in grid1:
        c = 0
        for col in row:
            draw.rectangle(xy=[c*50,r*50,(c+1)*50,(r+1)*50], fill=colorMap[np.abs(grid1[r][c])])
            c += 1
        r += 1
    for i in range(ncols):
        draw.line([(i+1)*50,0,(i+1)*50,height],fill="grey")
    for i in range(nrows):
        draw.line([0,(i+1)*50,width,(i+1)*50],fill="grey")

    
    # Grid 2
    grid2 = np.asarray(grid2)
    nrows = len(grid2[:,0])
    ncols = len(grid2[0,:])
    height, width = nrows*50, ncols*50
    image2 = Image.new(size=(width,height),mode='RGB',color=(255,255,255))
    draw = ImageDraw.Draw(image2)
    r = 0 
    for row in grid2:
        c = 0
        for col in row:
            draw.rectangle(xy=[c*50,r*50,(c+1)*50,(r+1)*50], fill=colorMap[np.abs(grid2[r][c])])
            c += 1
        r += 1
    for i in range(ncols):
        draw.line([(i+1)*50,0,(i+1)*50,height],fill="grey")
    for i in range(nrows):
        draw.line([0,(i+1)*50,width,(i+1)*50],fill="grey")
        
    # Displaying
    fig, ax = plt.subplots(1,2,figsize=(50,50))
    ax[0].imshow(image1)
    ax[0].axis("off")
    ax[1].imshow(image2)
    ax[1].axis("off")
```

```python
# checks if a portion of the grid is interesting, e.g., is more than 50% colored. this is a stand in, need a better 
# crtierion later
def colorCheck(rect):
    rect = np.asarray(rect)
    if np.count_nonzero(rect>0)/np.size(rect) >= 0.5:
        return True
    else:
        return False
```

```python
colorCheck(grid)
```

```python
# convolving the color check across the whole grid
def spatialConvolution(grid,rectSize,func=colorCheck):
    interestingRects = []
    grid = np.asarray(grid)
    nrows = len(grid[:,0])
    ncols = len(grid[0,:])
    for i in range(nrows):
        for j in range(ncols):
            if i + rectSize[0] > nrows or j + rectSize[1] > ncols:
                continue 
            elif func(grid[i:rectSize[0]+i,j:rectSize[1]+j]):
                interestingRects.append((i,j,rectSize))
    if interestingRects:
        return interestingRects
```

```python
# Trying out some codelets
def spatialCodelet(grid):
    interestingRects = []
    grid = np.asarray(grid)
    nrows = len(grid[:,0])
    ncols = len(grid[0,:])
    for i in range(nrows-1):
        for j in range(ncols-1):
            r = spatialConvolution(grid,(i+1,j+1))
            if r:
                interestingRects.append(spatialConvolution(grid,(i+1,j+1)))
    return interestingRects
```

```python
def diagCodelet(rect):
    rect = np.asarray(rect)
    if rect.shape[0] !=2 or rect.shape[1] != 2 or np.count_nonzero(rect>0)!=2:
        return False
    a = np.nonzero(rect)
    # i.e., both the rows and cols are different for the two colors
    if np.logical_xor(a[0][0],a[0][1]) and np.logical_xor(a[1][0],a[1][1]):
        return True
    else:
        return False
```

```python
def adjCodelet(rect):
    rect = np.asarray(rect)
    if rect.shape[0] !=2 or rect.shape[1] != 2 or np.count_nonzero(rect>0) < 2:
        return False
    a = np.nonzero(rect)
    # if there are 3 colors in a 4 unit rect, there's adjacency
    if len(a[0])==3:
        return True
    else:
        return False
```

```python
def subGrid(grid, rect):
    grid = np.asarray(grid)
    x = rect[0]
    y = rect[1]
    xSize = rect[2][0]
    ySize = rect[2][1]
    subGrid = grid[x:x+xSize,y:y+ySize]
    return subGrid
```
